```python
import matplotlib.pyplot as plt
```

```python
x1=[2,4,6,7]
y1=[34,51,29,45]
```

```python
plt.plot(x1,y1)
```

```python
plt.xlabel('Distance')
plt.ylabel('Time')
plt.plot(x1,y1)
```

```python
plt.grid(c='red')
plt.plot(x1,y1)
```

```python
x2=[12,15,16,17]
y2=[13,14,18,16]
plt.grid(c='green')
plt.xlabel('Distance')
plt.ylabel('Time')
plt.plot(x1,y1,label='cars')
plt.plot(x2,y2,label='bikes')
plt.legend()
```
